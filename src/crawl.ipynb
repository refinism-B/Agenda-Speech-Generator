# 司儀稿生成器

本程式使用LangChain自動爬蟲器進行資料擷取。

```python
from mod import A_LLM as lmp
from mod import B_text as tp
import pandas as pd
```

```python
# 輸入網址並取得網頁內容
url = "https://www.digitimes.com.tw/seminar/NTTDATA_20251212/"

web_content = lmp.load_web_info(url=url)
```

```python
# 選擇LLM模型並連線
model_name = 'gemini-3-flash-preview'

llm = lmp.connect_llm_model(model_name=model_name)
```

```python
# 建立prompt及相關資料
prompt_template = lmp.create_prompt_template()

json_parser, data = lmp.create_prompt_data(web_content=web_content)
```

```python
# 建立chain並啟動
llm_chain = prompt_template | llm

ai_message = llm_chain.invoke(data)

response = json_parser.invoke(ai_message)
input_token, out_token = lmp.get_tokens_info(ai_message=ai_message)
```

```python
# 預覽輸出結果
df = lmp.transform_to_df(response=response, key="agendas")

print(f"輸入token: {input_token}, 輸出token: {out_token}")
df
```

```python
# 轉換成純文字並存檔
text = tp.transform_to_text(result=response)

save_file = r"C:\Users\add41\Documents\Data_Engineer\Develop\tibame_LLM\data\text.txt"

tp.save_to_file(save_file=save_file, text=text)

print(f"輸入token: {input_token}, 輸出token: {out_token}")
```
